```python
#1 Dùng vòng lặp for hoặc while tính tổng các số từ 1 đến n với n <= 1000
n = int(input("Nhap 1 so nguyen duong n khong qua 1000:"))
s = 0
S = 0
k = 0
if 0 < n <= 1000:
    for i in range (n + 1):
        s += i
    while k <= n:
        S += k
        k += 1
    print("Tong la", s)
    print("Tong la", S)
else:
    print("Hay nhap so nguyen duong n phu hop")
```

```python
#2 Viết chương trình đến khi nhập được số nguyên dương thì thôi và kiểm tra có phải số nguyên tố không
while True:
    n = int(input("Nhap so nguyen duong n:"))
    if n <= 0:
        print("Hay nhap lai so nguyen duong!")
    else:
        break
if n >= 2:
    check = True
    for i in range (2 , int(n**0.5 + 1)):
        if n % i == 0:
            check = False
            break
    if check:
        print(n , "la so nguyen to")
    else:
        print(n , "khong la so nguyen to")
elif n == 1:
    print("1 khong phai so nguyen to")
```

```python
#3 Tính n! với 0 < n < 100:
n = int(input("Nhap 1 so nguyen duong n nho hon 100:"))
m = 1
M = 1
k = 1
if 0 < n < 100:
    for i in range (1 , n + 1):
        m *= i
    while k <= n:
        M *= k
        k += 1
    print("n! =" , m)
    print("n! =" , M)
else:
    print("Hay nhap gia tri phu hop cho n")
```

```python
#6 Tính tổng cấc số nguyên tố trong đoạn [a,b] (a <= b, a và b nguyên dương, mỗi dòng là một bộ test a,b)
while True:
    a , b = map(int,(input("Nhap so nguyen duong a va b cach nhau boi dau cach voi b >= a:").split()))
    s = 0
    if a <= b and a > 0:
        for i in range (a , b + 1):
            if i == 1:
                continue
            check = True
            for j in range (2 , int(i**0.5) + 1):
                if i % j == 0:
                    check = False
                    break
            if check == True:
                s += i
        print("Tong cac so nguyen to trong doan [a,b] la" , s)
    else:
        print("Hay nhap gia tri phu hop cho a, b!")
```

```python
#8 Nhập 1 số nguyên dương ban đầu, thực hiện phép toán n + reverse(n) đến khi thu được số đối xứng. In ra số đối xứng đó và số bước cần thiết
n = int(input("Nhap so nguyen duong n:"))
so_buoc = 0
if n > 0:
    def rev_n(n):
        rev_n = 0
        for i in range (len(str(n))):
            rev_n += (n % 10) * 10**(len(str(n))-1)
            n = n // 10
        return rev_n
    while n != rev_n(n):
        so_buoc += 1
        n = n + rev_n(n)
    print(f"So doi xung thu duoc la {n}, so buoc la {so_buoc}")
else:
    print("Hay nhap so nguyen duong phu hop")
```

```python
#10 Tìm số x trong đoạn [1,n] có độ dài dãy Collatz lớn nhất, nếu nhiều số cùng độ dài in số nhỏ nhất và độ dài lớn nhất
def Collatz(x):
    L = 1
    while x != 1:
        L += 1
        if x % 2 == 0:
            x /= 2
        else:
            x = 3*x + 1
    return L
n = int(input())
A = []
B = []
for x in range (1 , n + 1):
    A.append(Collatz(x))
for index , item in enumerate(A):
    if item == max(A):
        B.append(index)
print(f"So co do dai day Collatz lon nhat trong doan [1,{n}] la {min(B) + 1} , do dai la {max(A)}")
```

```python
#11 Đếm số ước chẵn của n với n không quá 10^6
n = int(input("Nhap so nguyen duong n:"))
N = 0
if 0 < n <= 1000000:
    for i in range (1 , n + 1):
        if n % (2*i) == 0:
            N += 1
    print(f"So uoc chan nguyen duong cua {n} la {N}")
else:
    print("Hay nhap gia tri phu hop cho n")
```

```python
#13 Kiểm tra xem 1 cặp số có phải cặp số thân thiết không
a , b = map(int,(input("Nhap 2 so nguyen duong cach nhau boi dau cach:").split()))
s1 = s2 = 0
for i in range (1 , a):
    if a % i == 0:
        s2 += i
for j in range (1 , b):
    if b % j == 0:
        s1 += j
if s1 == a and s2 == b:
    print("true")
else:
    print("false")
```

```python
#14 Tìm ước chung lớn nhất của m và n
m , n = map(int,(input("Nhap 2 so nguyen duong m va n cach nhau boi dau cach:").split()))
if m >= n > 0:
    for i in range (n , 0 , -1):
        if m % i == n % i == 0:
            print(f"Uoc chung lon nhat cua {m} va {n} la {i}")
            break
elif n > m > 0:
    for j in range (m , 0 , -1):
        if m % j == n % j == 0:
            print(f"Uoc chung lon nhat cua {m} va {n} la {j}")
            break
else:
    print("Hay nhap gia tri phu hop cho 2 so nguyen duong ban dau")
```
