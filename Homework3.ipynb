```python
#8 Tìm các ước chung của a và b
a , b = map(int,(input("Nhap 2 so nguyen duong cach nhau boi dau cach:").split()))
if a >= b > 0:
    for i in range (1 , b + 1):
        if a % i == b % i == 0:
            print(i , end = ' ')
elif b > a > 0:
    for j in range (1 , a + 1):
        if a % j == b % j == 0:
            print(j , end = ' ')
else:
    print("Hay nhap 2 so nguyen duong")
```

```python
#12 Tìm k sao cho tổng từ 1 đến k lớn nhất nhưng nhỏ hơn n
n = int(input("Nhap so nguyen duong n lon hon 1:"))
while n <= 1:
    n = int(input("Nhap so nguyen duong n lon hon 1:"))
else:
    s = 0
    k = 0
    while s < n:
        k += 1
        s += k
    print(f"k thoa man la {k - 1}")
```

```python
#13 Tìm n min sao cho tổng từ 1/1 đến 1/n là min, lớn hơn A
A = int(input("Nhap so nguyen duong:"))
while A <= 0:
    A = int(input("Nhap so nguyen duong:"))
else:
    s = 0
    k = 1
    while s <= A:
        s += 1/k
        k += 1
    print(f"k thoa man la {k - 1}")
```

```python
#14 Nhập 1 dãy số không dùng mảng, dừng khi nhập -1 và tìm min, max trong các số trên
lon_nhat = None
nho_nhat = None
while True:
    n = int(input("Nhap so nguyen (-1 de dung):"))
    if n == -1:
        break
    elif nho_nhat is None or n < nho_nhat:
        nho_nhat = n
    elif lon_nhat is None or n > lon_nhat:
        lon_nhat = n
print(f"So lon nhat la {lon_nhat} so nho nhat la {nho_nhat}")
```

```python
#15 Tìm số chữ số của số nguyên dương n
n = int(input("Nhap so nguyen duong n:"))
while n <= 0:
    n = int(input("Nhap so nguyen duong n:"))
else:
    scs = 1
    N = n
    print(f"So chu so cua {n} la {len(str(n))} chu so")
    while N // 10 != 0:
        scs += 1
        N //= 10
    print(f"So chu so cua {n} la {scs} chu so")
```

```python
#16 Đếm số chữ số chẵn và lẻ của số nguyên dương n
n = int(input("Nhap so nguyen duong n:"))
while n <= 0:
    n = int(input("Nhap so nguyen duong n:"))
else:
    a = str(n)
    chan = 0
    le = 0
    for i in range (len(a)):
        if int(a[i]) % 2 == 0:
            chan += 1
        else:
            le += 1
    print(f"{n} co {chan} chu so chan va {le} chu so le")
```

```python
#17 Tính tổng các chữ số của sô nguyên dương n
n = int(input("Nhap so nguyen duong n:"))
while n <= 0:
    n = int(input("Nhap so nguyen duong n:"))
else:
    s = 0
    N = n
    while N != 0:
        s += N % 10
        N //= 10
    print(f"Tong cac chu so cua {n} la {s}")
```

```python
#18 Kiểm tra 1 số nguyên dương có phải dạng 3^k không
n = int(input("Nhap so nguyen duong:"))
if n <= 0:
    print(f"{n} khong phai dang 3^k")
else:
    N = n
    while n % 3 == 0:
        n /= 3
    if n == 1:
        print(f"{N} la dang 3^k")
    else:
        print(f"{N} khong phai dang 3^k")
```

```python
#19 Tìm số trong dãy Fibonacci lớn nhất không quá A
A = int(input("Nhap so nguyen duong A:"))
m = 1
n = 1
while n < A:
    m = m + n
    n = m + n
if A < m:
    print(f"So gan nhat la {n-m}")
else:
    print(f"So gan nhat la {m}")
```

```python
#20 Đếm số chữ số nguyen tố của 1 số
n = int(input("Nhap so nguyen duong n:"))
a = str(n)
csnt = 0
for i in range (len(a)):
    if int(a[i]) == 2 or int(a[i]) == 3 or int(a[i]) == 5 or int(a[i]) == 7:
        csnt += 1
print(f"So chu so nguyen to cua {n} la {csnt} chu so")
```
