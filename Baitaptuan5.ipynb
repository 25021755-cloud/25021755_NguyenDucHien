```python
#1
a,b = map(int,input("Nhap 2 so nguyen a va b:").split())
def Max_Of_Two(a,b):
    if a >= b:
        return a
    else:
        return b
print("So lon hon la" , Max_Of_Two(a,b))
```

```python
#2
a,b = map(int,input("Nhap 2 so nguyen a va b:").split())
def swap(a,b):
    t = a
    a = b
    b = t
    return (a,b)
print(swap(a,b))
```

```python
#3
n = int(input("Nhap so nguyen duong:"))
def prime(n):
    for i in range (2 , int(n**0.5) + 1):
        if n % i == 0:
            return False
        return n >= 2
print(prime(n))
```

```python
#4
n = int(input("Nhap so nguyen duong:"))
def pf(n):
    s = 1
    for i in range (2 , int(n**0.5) + 1):
        if n % i == 0:
            s += (i + n/i)
            if i**2 == n:
                s -= i
    return s == n
print(pf(n))
```

```python
#5
so_nguyen = [int(x) for x in input().split()]
k = int(input("Nhap so nguyen k:"))
def find_k(so_nguyen):
    for i in range (len(so_nguyen)):
        if so_nguyen[i] == k:
            return i+1
    return -1
print(find_k(so_nguyen))
```

```python
#6
n = int(input("Nhap so nguyen n:")) 
def fac(n):
    m = 1
    if n == 0:
        return 1
    elif n < 0:
        return "Error"
    for i in range (1 , n + 1):
        m *= i
    return m
print("n!:" , fac(n))
```

```python
#7
num1 , num2 = map(float,(input("Nhap 2 so:").split()))
def caculation_(num1,num2):
    if num2 == 0:
        return "khong chia 0 duoc, hay nhap lai so thu 2"
    return round(num1 + num2 , 2), round(num1 - num2 , 2), round(num1 * num2 , 2) , round(num1/num2 , 2)
print(caculation_(num1,num2))
```

```python
#8
def d_Hamming(x,y):
    d = abs(len(str(bin(x))) - len(str(bin(y))))
    return d
x , y = map(int,(input("Nhap 2 so nguyen cach nhau dau cach").split()))
print(f"Khoang cach Hamming la {d_Hamming(x,y)}")
```

```python
#9
def tongchuso(n):
    s = 0
    a = str(n)
    for i in range (len(a)):
        s += int(a[i])
        n //= 10
    return s
n = int(input("Nhap so nguyen duong:"))
if n > 0:
    print(f"Tong cac chu so cua {n} la {tongchuso(n)}")
else:
    print("Nhap so nguyen duong")
```

```python
#10
def dangcau(c1,c2):
    anh1 = {}
    anh2 = {}
    if len(c1) != len(c2):
        return False
    for i , j in zip(c1,c2):
        if i in anh1:
            if anh1[i] != j:
                return False
        else:
            anh1[i] = j
        if j in anh2:
            if anh2[j] != i:
                return False
        else:
            anh2[j] = i
    return True
c1 = input("Nhap chuoi thu nhat:")
c2 = input("Nhap chuoi thu hai:")
if dangcau(c1,c2):
    print("true")
else:
    print("false")
```
