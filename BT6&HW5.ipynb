```python
#5
A = list(map(int,(input().split())))
p = 0
z = 0
n = 0
d = {}
for x in A:
    if x > 0:
        p += 1
    elif x < 0:
        n += 1
    else:
        z += 1
    d['positives'] = p
    d['negatives'] = n
    d['zeros'] = z
print(d)
```

```python
#9
a1 = eval(input())
a2 = eval(input())
d = {}
for k1 , v1 in a1.items():
    d[k1] = [v1]
for k2 , v2 in a2.items():
    if k2 in d:
        d[k2] = d[k2] + [v2]
    else:
        d[k2] = [v2]
print(dict(sorted(d.items())))
```

```python
#10
sn = list(map(int,(input().split())))
k = int(input())
result_set = set()
for i in range(len(sn)):
    for j in range(i + 1, len(sn)):
        if sn[i] + sn[j] == k:
            result_set.add((sn[i], sn[j]))
result = sorted(result_set)
print(result)
```

```python
#12
sn_list = list(map(int,(input().split())))
sn_dnhat = set(sn_list)
list_dnhat = list(sn_dnhat)
xhien = []
chisodonglonnhat = []
xhienmax = []
for x in sn_dnhat:
    n = 0
    for k in sn_list:
        if k == x:
            n += 1
    xhien.append(n)
for i in range (len(xhien)):
    if xhien[i] == max(xhien):
        chisodonglonnhat.append(i)
        xhienmax.append(list_dnhat[i])
print(min(xhienmax))
```

```python
#14
A = list(map(int,(input().split())))
B = list(map(int,(input().split())))
giao = []
giaodnhat = set()
kq = []
for x in A:
    for k in B:
        if x == k:
            giao.append(x)
for a in giao:
    if a not in giaodnhat:
        giaodnhat.add(a)
        kq.append(a)
print(kq)
```

```python
#24
n = list(map(int,(input().split())))
count0 = 0
for i in range (len(n) - 1 , -1 , -1):
    if n[i] == 0:
        count0 += 1
        n.remove(0)
n.reverse()
for j in range (count0):
    n.append(0)
n.reverse()
print(n)
```
